### lets see if abrupt boundaries contribute to the "bright cross" in the Fourier spectrum

```python
# imports

import scipy.misc
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading and plotting an image
img = scipy.misc.ascent()
plt.imshow(img, cmap="gray")
```

```python
# doing 2d Fourier Transform and plotting the absolute value (the cross is visible)
F = np.fft.fftshift(np.fft.fft2(img))
plt.imshow(np.abs(F), cmap="gray", vmax = 10000)
```

```python
# calculating and plotting a gaussian
sigma = 100
sx,sy = img.shape
mux, muy = (.5*sx,.5*sy)
x = np.linspace(0,sx,sx)
y = np.linspace(0,sy,sy)
xx,yy = np.meshgrid(x,y)
ga = np.exp(-((xx-mux)**2 + (yy-muy)**2)/(2*sigma**2))

plt.imshow(ga)
```

```python
# applying gaussian as a window
plt.imshow(img*ga)
```

```python
# calculating and plotting the Fourier spectrum of the windowed image
# IMPORTANT: to properly compare the spectra of the image with and without windowing 
# we need to properly normalize them first. But just from the visual analysis and one can probably see
# that the cross disapears
Fw = np.fft.fftshift(np.fft.fft2(img*ga))
plt.imshow(np.abs(Fw), cmap="gray", vmax = 1000)
```
