```python
import tensorflow as tf
import numpy as np
```

```python
a = tf.constant(np.arange(12,dtype=np.int32), shape=[2,2,3])
b = tf.constant(10*np.arange(12, dtype=np.int32), shape=[2,3,2])
c = tf.constant(np.ones(2,dtype=np.int32), shape=[10,1,2])
d = tf.constant(np.ones((10,2,6), dtype=np.int32))
```

```python
sess = tf.InteractiveSession()
```

```python
sess.run(a)
```

```python
sess.run(b)
```

```python
sess.run(c)
```

```python
sess.run(d)
```

```python
sess.run(tf.matmul(c,d))
```

```python
ff = np.arange(15)
```

```python
x,y,z = (25,ff.shape[0],35)
```

```python
T = tf.constant(np.arange(2*z*y ,dtype=np.int32), shape=[2*z,y])
TT = tf.constant(np.arange(2*z*y ,dtype=np.int32), shape=[y,z*2])
```

```python
f = tf.constant(np.stack(z*[ff],axis=0).astype(np.int32), shape=[z,1,y])
```

```python
T.get_shape().as_list()
```

```python
f.get_shape().as_list()
```

```python
m = tf.matmul(TT,T)
```

```python
sess.run(T)
```

```python
sess.run(f)
```
