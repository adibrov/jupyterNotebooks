```python
datasetDirectory = "../../data/franz/"
timePointDirectory = "WT_25deg_160413-03/"
```

```python
import tifffile
tile = tifffile.imread(datasetDirectory+timePointDirectory+"P5.tif")
hm = tifffile.imread(datasetDirectory+timePointDirectory+"P5_proj_r20_d2_t50_layer0_interpolatedHeightMap.tif")
```

```python
tile.shape
```

```python
x,y = hm.shape
```

```python
hm[0,0]
```

```python
import numpy as np
proj = np.zeros((x,y))
```

```python
for i in range(x):
    for j in range(y):
        proj[i,j] = tile[hm[i,j],i,j]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,10
plt.imshow(proj)
```

```python
tifffile.imsave(datasetDirectory+timePointDirectory+"manualProj.tif", proj.astype(np.uint16))
```
